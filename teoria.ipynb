# ¿Que reciben los ejes?
Sí, en la mayoría de las bibliotecas de visualización de datos como Matplotlib y Seaborn, puedes pasarle varios tipos de estructuras de datos a los ejes "x" e "y" para representarlos en un gráfico. Algunos de los tipos de datos que generalmente puedes utilizar son:

1. Listas: Las listas son una de las estructuras de datos más comunes en Python y se pueden utilizar para representar datos en los ejes "x" e "y" de un gráfico. Por ejemplo:
   ```python
   x = [1, 2, 3, 4, 5]
   y = [10, 20, 15, 25, 30]
   ```

2. Tuplas: Las tuplas son similares a las listas, pero son inmutables. También se pueden utilizar para representar datos en los ejes "x" e "y" del gráfico. Por ejemplo:
   ```python
   x = (1, 2, 3, 4, 5)
   y = (10, 20, 15, 25, 30)
   ```

3. Data Frame de Pandas: Si estás utilizando la biblioteca Pandas, puedes utilizar un df de Pandas para representar datos en los ejes "x" e "y" del gráfico. Una serie es una estructura de datos unidimensional de Pandas que puede contener datos numéricos o categóricos. Por ejemplo:
   ```python
   import pandas as pd
   
   data = {'x': [1, 2, 3, 4, 5], 'y': [10, 20, 15, 25, 30]}
   df = pd.DataFrame(data)
   x = df['x']
   y = df['y']
   ```

4. Numpy Arrays: Si estás utilizando la biblioteca Numpy, puedes utilizar arrays de Numpy para representar datos en los ejes "x" e "y" del gráfico. Los arrays de Numpy pueden contener datos numéricos. Por ejemplo:
   ```python
   import numpy as np
   
   x = np.array([1, 2, 3, 4, 5])
   y = np.array([10, 20, 15, 25, 30])
   ```


5. Fechas y Horas: Puedes pasar objetos de fecha y hora para representar datos en el eje "x" o "y" cuando estés trabajando con series temporales o datos que tienen un componente de tiempo. Por ejemplo:
   ```python
   import pandas as pd
   
   dates = pd.date_range(start='2023-01-01', periods=5, freq='D')
   values = [10, 20, 15, 25, 30]
   ```

6. Cadenas de Texto: En ciertos tipos de gráficos, como gráficos de barras o gráficos de dispersión, puedes utilizar cadenas de texto para representar datos en el eje "x" o "y". Por ejemplo, puedes mostrar nombres de ciudades o etiquetas específicas. Asegúrate de que las cadenas de texto sean únicas y no repetidas. Por ejemplo:
   ```python
   x = ['Ciudad A', 'Ciudad B', 'Ciudad C', 'Ciudad D']
   y = [100, 150, 120, 200]
   ```

7. Funciones: En algunos casos, puedes pasar una función que genere datos para los ejes "x" o "y". Esto puede ser útil cuando deseas personalizar la representación de los datos o crear gráficos específicos. Por ejemplo:
   ```python
   import numpy as np
   
   def custom_function(x):
       return np.sin(x)
   
   x = np.linspace(0, 2*np.pi, 100)
   y = custom_function(x)
   ```

8. Diccionarios: En ciertos escenarios, como al trabajar con gráficos de barras agrupadas o gráficos de barras apiladas, puedes utilizar diccionarios para representar datos en los ejes "x" o "y". Los diccionarios pueden contener etiquetas de grupos y subgrupos, lo que facilita la visualización de datos. Por ejemplo:
   ```python
   data = {
       'Grupo A': [10, 20, 15],
       'Grupo B': [30, 25, 35],
       'Grupo C': [5, 10, 8]
   }
   ```

## plt.scatter(): 
Este método se utiliza para crear gráficos de dispersión. Toma dos conjuntos de datos (valores para los ejes x e y) y representa cada punto como un marcador en el gráfico. Es útil para visualizar relaciones entre dos variables y detectar patrones o agrupamientos.

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.scatter(x, y)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Dispersión')
plt.show()
````

## plt.step(): 
Este método crea gráficos de escalones, donde los puntos se unen por segmentos horizontales y verticales. Es útil para representar funciones discontinuas o datos discretos que cambian solo en ciertos puntos.

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.step(x, y)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Escalones')
plt.show()
````

## plt.stem():
Este método crea gráficos de tallo y hojas, donde se muestra la distribución de datos. Es útil para visualizar la magnitud y ubicación de cada valor en un conjunto de datos.



````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.stem(x, y)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Tallo y Hojas')
plt.show()
````

## plt.fill_between(): 
Este método se utiliza para rellenar el área entre dos curvas o líneas. Es útil para resaltar áreas bajo o entre curvas y para visualizar incertidumbres en datos.

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [10, 20, 15, 25, 30]
y2 = [5, 15, 10, 20, 25]

plt.plot(x, y1, label='Línea 1')
plt.plot(x, y2, label='Línea 2')

plt.fill_between(x, y1, y2, color='gray', alpha=0.5)

plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico con fill_between')
plt.legend()
plt.show()
````

## plt.errorbar():
Este método se utiliza para crear gráficos de líneas con barras de error en los puntos de datos. Es útil para visualizar la incertidumbre o variabilidad en los datos.

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]
error = [1, 2, 1, 3, 2]

plt.errorbar(x, y, yerr=error, fmt='o', capsize=4)

plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico con barras de error')
plt.show()
````

# Estilos y personalizacion de las graficas mediante diferentes parametros
Matplotlib, puedes usar `mpl.style.use()` para aplicar diferentes estilos predefinidos y hacer que tus gráficos luzcan más atractivos y profesionales.

Aparte del estilo 'bmh', Matplotlib ofrece una variedad de estilos incorporados que puedes utilizar. Algunos de ellos son:

1. `'ggplot'`: Emula el estilo del paquete ggplot de R, popular por su apariencia limpia y moderna.

2. `'seaborn'`: Emula el estilo de Seaborn, que es una biblioteca de visualización de datos basada en Matplotlib.

3. `'fivethirtyeight'`: Emula el estilo del sitio web FiveThirtyEight, conocido por sus gráficos de datos y políticos.

4. `'dark_background'`: Fondo oscuro con líneas y texto en blanco, útil para resaltar datos en gráficos.

5. `'classic'`: Estilo clásico de Matplotlib, que se asemeja al aspecto tradicional de las visualizaciones de Matplotlib.

Estos son solo algunos ejemplos de los estilos incorporados en Matplotlib. Puedes consultar la lista completa de estilos disponibles ejecutando el siguiente código:

```python
import matplotlib.pyplot as plt

print(plt.style.available)
```

Además de los estilos predefinidos, también puedes personalizar el estilo de tus gráficos ajustando diferentes parámetros. Puedes cambiar cosas como el tamaño de la fuente, el tamaño de los márgenes, los colores predeterminados y mucho más. Para ello, puedes utilizar el método `plt.rcParams` para establecer los parámetros deseados. Por ejemplo:

```python
import matplotlib.pyplot as plt

# Cambiar el tamaño de fuente predeterminado
plt.rcParams['font.size'] = 12

# Cambiar los colores predeterminados para las líneas en los gráficos
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['lines.color'] = 'r'

# Cambiar el tamaño de las figuras predeterminado
plt.rcParams['figure.figsize'] = (8, 6)

# Cambiar el tamaño de los márgenes
plt.rcParams['figure.subplot.left'] = 0.1
plt.rcParams['figure.subplot.right'] = 0.9
plt.rcParams['figure.subplot.bottom'] = 0.1
plt.rcParams['figure.subplot.top'] = 0.9

# Y así sucesivamente, hay muchos más parámetros que puedes ajustar según tus necesidades.
```

Puedes combinar diferentes estilos y ajustes de parámetros para crear gráficos personalizados y adaptados a tus preferencias y necesidades específicas. Experimenta con los estilos y parámetros disponibles para obtener el aspecto deseado en tus visualizaciones.
