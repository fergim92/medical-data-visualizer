# ¿Que reciben los ejes?
Sí, en la mayoría de las bibliotecas de visualización de datos como Matplotlib y Seaborn, puedes pasarle varios tipos de estructuras de datos a los ejes "x" e "y" para representarlos en un gráfico. Algunos de los tipos de datos que generalmente puedes utilizar son:

1. Listas: Las listas son una de las estructuras de datos más comunes en Python y se pueden utilizar para representar datos en los ejes "x" e "y" de un gráfico. Por ejemplo:
   ```python
   x = [1, 2, 3, 4, 5]
   y = [10, 20, 15, 25, 30]
   ```

2. Tuplas: Las tuplas son similares a las listas, pero son inmutables. También se pueden utilizar para representar datos en los ejes "x" e "y" del gráfico. Por ejemplo:
   ```python
   x = (1, 2, 3, 4, 5)
   y = (10, 20, 15, 25, 30)
   ```

3. Data Frame de Pandas: Si estás utilizando la biblioteca Pandas, puedes utilizar un df de Pandas para representar datos en los ejes "x" e "y" del gráfico. Una serie es una estructura de datos unidimensional de Pandas que puede contener datos numéricos o categóricos. Por ejemplo:
   ```python
   import pandas as pd
   
   data = {'x': [1, 2, 3, 4, 5], 'y': [10, 20, 15, 25, 30]}
   df = pd.DataFrame(data)
   x = df['x']
   y = df['y']
   ```

4. Numpy Arrays: Si estás utilizando la biblioteca Numpy, puedes utilizar arrays de Numpy para representar datos en los ejes "x" e "y" del gráfico. Los arrays de Numpy pueden contener datos numéricos. Por ejemplo:
   ```python
   import numpy as np
   
   x = np.array([1, 2, 3, 4, 5])
   y = np.array([10, 20, 15, 25, 30])
   ```


5. Fechas y Horas: Puedes pasar objetos de fecha y hora para representar datos en el eje "x" o "y" cuando estés trabajando con series temporales o datos que tienen un componente de tiempo. Por ejemplo:
   ```python
   import pandas as pd
   
   dates = pd.date_range(start='2023-01-01', periods=5, freq='D')
   values = [10, 20, 15, 25, 30]
   ```

6. Cadenas de Texto: En ciertos tipos de gráficos, como gráficos de barras o gráficos de dispersión, puedes utilizar cadenas de texto para representar datos en el eje "x" o "y". Por ejemplo, puedes mostrar nombres de ciudades o etiquetas específicas. Asegúrate de que las cadenas de texto sean únicas y no repetidas. Por ejemplo:
   ```python
   x = ['Ciudad A', 'Ciudad B', 'Ciudad C', 'Ciudad D']
   y = [100, 150, 120, 200]
   ```

7. Funciones: En algunos casos, puedes pasar una función que genere datos para los ejes "x" o "y". Esto puede ser útil cuando deseas personalizar la representación de los datos o crear gráficos específicos. Por ejemplo:
   ```python
   import numpy as np
   
   def custom_function(x):
       return np.sin(x)
   
   x = np.linspace(0, 2*np.pi, 100)
   y = custom_function(x)
   ```

8. Diccionarios: En ciertos escenarios, como al trabajar con gráficos de barras agrupadas o gráficos de barras apiladas, puedes utilizar diccionarios para representar datos en los ejes "x" o "y". Los diccionarios pueden contener etiquetas de grupos y subgrupos, lo que facilita la visualización de datos. Por ejemplo:
   ```python
   data = {
       'Grupo A': [10, 20, 15],
       'Grupo B': [30, 25, 35],
       'Grupo C': [5, 10, 8]
   }
   ```

## plt.scatter(): dispersión
Este método se utiliza para crear gráficos de dispersión. Toma dos conjuntos de datos (valores para los ejes x e y) y representa cada punto como un marcador en el gráfico. Es útil para visualizar relaciones entre dos variables y detectar patrones o agrupamientos.

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.scatter(x, y)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Dispersión')
plt.show()
````

## plt.step(): escalones
Este método crea gráficos de escalones, donde los puntos se unen por segmentos horizontales y verticales. Es útil para representar funciones discontinuas o datos discretos que cambian solo en ciertos puntos.

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.step(x, y)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Escalones')
plt.show()
````

## plt.stem(): tallo y hojas
Este método crea gráficos de tallo y hojas, donde se muestra la distribución de datos. Es útil para visualizar la magnitud y ubicación de cada valor en un conjunto de datos.



````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.stem(x, y)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Tallo y Hojas')
plt.show()
````

## plt.fill_between():  área entre dos curvas o líneas
Este método se utiliza para rellenar el área entre dos curvas o líneas. Es útil para resaltar áreas bajo o entre curvas y para visualizar incertidumbres en datos.

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [10, 20, 15, 25, 30]
y2 = [5, 15, 10, 20, 25]

plt.plot(x, y1, label='Línea 1')
plt.plot(x, y2, label='Línea 2')

plt.fill_between(x, y1, y2, color='gray', alpha=0.5)

plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico con fill_between')
plt.legend()
plt.show()
````

## plt.errorbar(): barras de error
Este método se utiliza para crear gráficos de líneas con barras de error en los puntos de datos. Es útil para visualizar la incertidumbre o variabilidad en los datos.

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]
error = [1, 2, 1, 3, 2]

plt.errorbar(x, y, yerr=error, fmt='o', capsize=4)

plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico con barras de error')
plt.show()
````

## Ubicacion de la leyenda en el grafico(la info de la grafica)

En Matplotlib, el parámetro `loc` se utiliza para especificar la ubicación de la leyenda en el gráfico. La leyenda es un cuadro que identifica cada una de las series o elementos representados en el gráfico, etiquetando cada uno con un nombre o etiqueta descriptiva. El argumento `loc` acepta diferentes cadenas de texto o códigos numéricos para controlar la ubicación de la leyenda en el gráfico.

**Agregar leyendas**
```python
ax.legend()  
```

Aquí están algunos de los valores más comunes que puedes utilizar para el parámetro `loc`:

1. `'best'` o `0`: Matplotlib intenta encontrar automáticamente la mejor ubicación para la leyenda, minimizando el solapamiento con los datos del gráfico.

2. `'upper right'` o `1`: Coloca la leyenda en la esquina superior derecha del gráfico.

3. `'upper left'` o `2`: Coloca la leyenda en la esquina superior izquierda del gráfico.

4. `'lower left'` o `3`: Coloca la leyenda en la esquina inferior izquierda del gráfico.

5. `'lower right'` o `4`: Coloca la leyenda en la esquina inferior derecha del gráfico.

6. `'right'` o `5`: Coloca la leyenda en el lado derecho del gráfico, centrada verticalmente.

7. `'center left'` o `6`: Coloca la leyenda en el lado izquierdo del gráfico, centrada verticalmente.

8. `'center right'` o `7`: Coloca la leyenda en el lado derecho del gráfico, centrada verticalmente.

9. `'lower center'` o `8`: Coloca la leyenda en la parte inferior del gráfico, centrada horizontalmente.

10. `'upper center'` o `9`: Coloca la leyenda en la parte superior del gráfico, centrada horizontalmente.

11. `'center'` o `10`: Coloca la leyenda en el centro del gráfico.

También puedes utilizar tuplas de coordenadas (x, y) para ubicar manualmente la leyenda en una posición específica dentro del gráfico. Por ejemplo:

```python
ax.legend(loc=(0.7, 0.7))  # Ubica la leyenda en la posición (0.7, 0.7) del gráfico
```

Cabe destacar que las ubicaciones exactas pueden variar ligeramente dependiendo del tamaño y proporciones del gráfico, así que es recomendable ajustar la ubicación según tus necesidades específicas. Experimenta con diferentes valores de `loc` para encontrar la ubicación de leyenda más adecuada para tu gráfico.

Tipo de gráfico que muestra información como una serie de puntos de datos conectados por segmentos de línea recta. 
Se utilizan generalmente para visualizar el movimiento direccional de uno o más datos a lo largo del tiempo.

````python
# El anterior codigo es equivalente a este
fig, ax = plt.subplots()
ax.plot([1, 5, 2, 4, 8, 9, 2])

#¿Porque?
'''
Porque :La razón por la cual las dos líneas de código son equivalentes es debido al comportamiento predeterminado 
de Matplotlib cuando solo se proporciona una lista de valores para `ax.plot()` sin especificar el eje x.

En la primera línea de código:
```python
fig, ax = plt.subplots()
ax.plot([1, 5, 2, 4, 8, 9, 2])
```
Matplotlib toma la lista `[1, 5, 2, 4, 8, 9, 2]` como los valores para el eje y. 
Sin embargo, como no se proporcionaron valores para el eje x, Matplotlib asume que el eje x corresponde a 
los índices de los elementos en la lista `[1, 5, 2, 4, 8, 9, 2]`. Los índices en este caso son `[0, 1, 2, 3, 4, 5, 6]`
 ya que la lista comienza en el índice 0 y termina en el índice 6.

En la segunda línea de código:
```python
fig, ax = plt.subplots()
ax.plot([0, 1, 2, 3, 4, 5, 6], [1, 5, 2, 4, 8, 9, 2])
```
Aquí, se proporcionan explícitamente los valores tanto para el eje x como para el eje y. En este caso, los valores
 `[0, 1, 2, 3, 4, 5, 6]` se utilizan como el eje x, y los valores `[1, 5, 2, 4, 8, 9, 2]` se utilizan como el eje y.

Entonces, ambas líneas de código generan el mismo gráfico de línea. Si no necesitas controlar los valores específicos 
del eje x y estás satisfecho con los índices de los datos como valores para el eje x, la primera línea de código es más 
concisa y se utiliza comúnmente cuando no se requieren valores de eje x personalizados. Si deseas tener más control sobre los valores del eje x, entonces puedes usar la segunda línea de código y proporcionar los valores de eje x y eje y por separado.

'''
````

## hist() Grafico Histograma(1 variable)
Un histograma es una gráfica de la distribución de frecuencia de una matriz numérica dividiéndola en pequeños contenedores de igual tamaño (bins).

**EL EJE X ES DE SOLO UNA VARIABLE** el eje x son rangos no categorias

Los bins usualmente se calculan por medio de la fórmula de sturges


La altura de cada barra representa la proporción o cantidad de los distintos valores de una variable numérica.
Requiere clasificar a los datos en intervalos de clase.
Permiten comparar la frecuencia relativa o absoluta de cada intervalo.
Se construyen con ax.hist, que recibe como parámetro:
El arreglo de valores. 
bins, que representa la cantidad de intervalos a construir

Observaciones
La forma del histograma depende del número de intervalos de clase que pasemos al parámetro bins.
En el ejemplo se representó la frecuencia absoluta de los intervalos.
También se puede representar la frecuencia relativa de cada intervalo o el porcentaje respecto del total.


````python
#Aplanemos los valores del Data Frame con el método flatten
# FLATTEN: meter todos los valores en una sola lista
fig, ax = plt.subplots( figsize=(12,4)) #definir el lienzo u hoja en blanco

#espacio para conf del grafico
ax.set_title('Histograma de precipitaciones')
ax.set_xlabel('Intervalos')
ax.set_ylabel('Frecuencia absoluta - Conteo')

#espacio para ejecucion de la grafica
ax.hist(df_lluvias.values.flatten(), color='green')

#leyenda
````

# Estilos y personalizacion de las graficas mediante diferentes parametros
Matplotlib, puedes usar `mpl.style.use()` para aplicar diferentes estilos predefinidos y hacer que tus gráficos luzcan más atractivos y profesionales.

Aparte del estilo 'bmh', Matplotlib ofrece una variedad de estilos incorporados que puedes utilizar. Algunos de ellos son:

1. `'ggplot'`: Emula el estilo del paquete ggplot de R, popular por su apariencia limpia y moderna.

2. `'seaborn'`: Emula el estilo de Seaborn, que es una biblioteca de visualización de datos basada en Matplotlib.

3. `'fivethirtyeight'`: Emula el estilo del sitio web FiveThirtyEight, conocido por sus gráficos de datos y políticos.

4. `'dark_background'`: Fondo oscuro con líneas y texto en blanco, útil para resaltar datos en gráficos.

5. `'classic'`: Estilo clásico de Matplotlib, que se asemeja al aspecto tradicional de las visualizaciones de Matplotlib.

Estos son solo algunos ejemplos de los estilos incorporados en Matplotlib. Puedes consultar la lista completa de estilos disponibles ejecutando el siguiente código:

```python
import matplotlib.pyplot as plt

print(plt.style.available)
```

Además de los estilos predefinidos, también puedes personalizar el estilo de tus gráficos ajustando diferentes parámetros. Puedes cambiar cosas como el tamaño de la fuente, el tamaño de los márgenes, los colores predeterminados y mucho más. Para ello, puedes utilizar el método `plt.rcParams` para establecer los parámetros deseados. Por ejemplo:

```python
import matplotlib.pyplot as plt

# Cambiar el tamaño de fuente predeterminado
plt.rcParams['font.size'] = 12

# Cambiar los colores predeterminados para las líneas en los gráficos
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['lines.color'] = 'r'

# Cambiar el tamaño de las figuras predeterminado
plt.rcParams['figure.figsize'] = (8, 6)

# Cambiar el tamaño de los márgenes
plt.rcParams['figure.subplot.left'] = 0.1
plt.rcParams['figure.subplot.right'] = 0.9
plt.rcParams['figure.subplot.bottom'] = 0.1
plt.rcParams['figure.subplot.top'] = 0.9

# Y así sucesivamente, hay muchos más parámetros que puedes ajustar según tus necesidades.
```

Puedes combinar diferentes estilos y ajustes de parámetros para crear gráficos personalizados y adaptados a tus preferencias y necesidades específicas. Experimenta con los estilos y parámetros disponibles para obtener el aspecto deseado en tus visualizaciones.

# SEABORN


1. Relational plots:
Los gráficos relacionales se utilizan para visualizar la relación entre dos variables numéricas. Un ejemplo común de un gráfico relacional es el diagrama de dispersión (scatter plot). En este tipo de gráfico, cada punto representa una observación con valores en ambos ejes, y la posición del punto muestra cómo se relacionan las dos variables. Estos gráficos son útiles para identificar patrones, tendencias y la fuerza de la relación entre las variables. Se utilizan especialmente cuando se quiere explorar si existe alguna correlación entre las variables y cómo se distribuyen conjuntamente.

2. Distribution plots:
Los gráficos de distribución se utilizan para visualizar la distribución de una variable numérica o un conjunto de variables numéricas. Algunos ejemplos comunes de gráficos de distribución son el histograma, el gráfico de densidad y el diagrama de cajas (box plot). El histograma muestra la frecuencia de los valores en diferentes rangos, el gráfico de densidad muestra la distribución continua de probabilidad y el diagrama de cajas muestra la mediana, cuartiles y valores atípicos. Estos gráficos son útiles para entender cómo se distribuyen los datos y si hay valores atípicos o sesgos en los datos.

3. Categorical plots:
Los gráficos categóricos se utilizan para visualizar la relación entre variables categóricas o para mostrar la distribución de datos categóricos. Algunos ejemplos comunes de gráficos categóricos son los gráficos de barras (bar plot), los gráficos de conteo (count plot) y los gráficos de pastel (pie chart). Estos gráficos son útiles cuando queremos comparar las frecuencias o proporciones entre diferentes categorías o grupos.

4. Regression plots:
Los gráficos de regresión se utilizan para visualizar la relación entre dos variables numéricas y ajustar una línea o curva de regresión a los datos. Un ejemplo común de un gráfico de regresión es el diagrama de dispersión con una línea de regresión ajustada. Este tipo de gráficos es especialmente útil cuando se quiere identificar la relación lineal (o no lineal) entre dos variables y ver cómo se ajusta el modelo de regresión a los datos.

5. Matrix plots:
Los gráficos de matriz, también conocidos como matrices de dispersión (scatterplot matrix) o matrices de correlación, se utilizan para visualizar las relaciones entre múltiples variables numéricas. Cada celda de la matriz representa un diagrama de dispersión entre dos variables, y la diagonal principal generalmente muestra las distribuciones de cada variable. Estos gráficos son útiles cuando se quiere explorar patrones de correlación o tendencias entre múltiples variables a la vez.

En resumen, cada tipo de gráfico tiene su función específica y se utiliza para visualizar diferentes aspectos de los datos, ya sea para explorar relaciones entre variables numéricas, distribuciones de datos, comparaciones categóricas o análisis de regresión. La elección del gráfico adecuado depende de los objetivos específicos del análisis y la naturaleza de los datos que se estén estudiando.
